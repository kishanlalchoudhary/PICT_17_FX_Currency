```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import seaborn as sns
```

# read the data from csv format to dataframes

```python
df1 = pd.read_csv("Exchange_Rate_Report_2012.csv")
df2 = pd.read_csv("Exchange_Rate_Report_2013.csv")
df3 = pd.read_csv("Exchange_Rate_Report_2014.csv")
df4 = pd.read_csv("Exchange_Rate_Report_2015.csv")
df5 = pd.read_csv("Exchange_Rate_Report_2016.csv")
df6 = pd.read_csv("Exchange_Rate_Report_2017.csv")
df7 = pd.read_csv("Exchange_Rate_Report_2018.csv")
df8 = pd.read_csv("Exchange_Rate_Report_2019.csv")
df9 = pd.read_csv("Exchange_Rate_Report_2020.csv")
df10 = pd.read_csv("Exchange_Rate_Report_2021.csv")
df11 = pd.read_csv("Exchange_Rate_Report_2022.csv")
     
```

# store all dates columns

```python

fc1 = df1.pop("Date")
fc2 = df2.pop("Date")
fc3 = df3.pop("Date")
fc4 = df4.pop("Date")
fc5 = df5.pop("Date")
fc6 = df6.pop("Date")
fc7 = df7.pop("Date")
fc8 = df8.pop("Date")
fc9 = df9.pop("Date")
fc10 = df10.pop("Date")
fc11 = df11.pop("Date")
     
```

# sort the data in alphabetical order of currencies

```python
df1 = df1.reindex(sorted(df1.columns), axis=1)
df2 = df2.reindex(sorted(df2.columns), axis=1)
df3 = df3.reindex(sorted(df3.columns), axis=1)
df4 = df4.reindex(sorted(df4.columns), axis=1)
df5 = df5.reindex(sorted(df5.columns), axis=1)
df6 = df6.reindex(sorted(df6.columns), axis=1)
df7 = df7.reindex(sorted(df7.columns), axis=1)
df8 = df8.reindex(sorted(df8.columns), axis=1)
df9 = df9.reindex(sorted(df9.columns), axis=1)
df10 = df10.reindex(sorted(df10.columns), axis=1)
df11 = df11.reindex(sorted(df11.columns), axis=1)
```

# insert the date column which we removed earlier

# concatenate all data frames

```python
frames = [df1, df2, df3,df4,df5,df6,df7,df8,df9,df10,df11]
dt = pd.concat(frames)
```

# represent currencies in their USD Codes

```python
new_cols = ['Date']
for name in dt.columns[1:]:
  loc = name.index("(")
  new_cols.append(name[loc + 1:loc + 4])
dt.columns = new_cols
print(dt.columns)
     
```

# displaying last 7 values of Date Column

```python

dt["Date"].tail(7)
```

```python
import datetime
# Creating an empty list to store modified date values
new_dates = []
# The format of the date strings in the "Date" column of the DataFrame (e.g., '25-Jan-21')
format_str = '%d-%b-%y' # This format represents day-month-year
# Iterating through each date string in the "Date" column of the DataFrame
for date_str in dt["Date"]: 
    # Convert the date string to a datetime object using the specified format
  datetime_obj = datetime.datetime.strptime(date_str, format_str)
    # Convert the datetime object back to a string in the format '%Y-%m-%d' (year-month-day)
  date_obj = datetime.datetime.strptime(date_str, format_str).strftime("%Y-%m-%d")
   # Append the modified date string to the new_dates list
  new_dates.append(date_obj)
# Replace the values in the "Date" column of the DataFrame with the modified dates
dt["Date"] = new_dates
```
